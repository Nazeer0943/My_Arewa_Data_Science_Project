```python
#we need to import the necessary libraries for the plots and the diagram, such as matplotib, numpy and os.

import os
import numpy as np
import matplotlib.pyplot as plt

# This is make plots directory if it does not exist
os.makedirs("plots", exist_ok=True)

# sample data
categories = ["Apples", "Samsung", "Infinix", "Redmi"]
sales = np.random.randint(50, 200, size=len(categories))

# We are plotting the randomly geenrated data and the sample data using a Bar Chart
plt.bar(categories, sales, color="skyblue")
plt.title("Sales by phones Type")
plt.xlabel("phones")
plt.ylabel("Sales")
plt.savefig("plots/bar_chart.png")  
plt.show()
```
